```python
import matplotlib.pyplot as plt
def inv_annealing_poly(start, end, pct:float, degree, **kwargs):
    "Helper function for `inv_anneal_poly`."
    return start + (end - start) * (pct)**degree

def do_annealing_poly(start, end, pct:float, degree):
    "Helper function for `anneal_poly`."
    return end + (start-end) * (1-pct)**degree

degree = 3
a_ = []
b_ = []
for i in range(1000):
    #a_.append(inv_annealing_poly(1, 0, i/1000, degree))
    a_.append(inv_annealing_poly(0, 1, i/1000, degree))
    #a_.append(do_annealing_poly(0, 1, i/1000, degree))
    b_.append(do_annealing_poly(1, 0, i/1000, degree))
plt.plot(a_)
plt.plot(b_)
plt.show()
```
